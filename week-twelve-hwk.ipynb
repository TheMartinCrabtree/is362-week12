Task:
Choose one of the New York Times APIs, construct an interface in Python to read in the JSON data, and transform it into a pandas DataFrame:

nyt-books-bestseller-overview.json has been downloaded from the NYT Books/Best Sellers list API. 

The endpoint URL for the current NYT best sellers list for hardcover fiction:
https://api.nytimes.com/svc/books/v3/lists/current/hardcover-fiction.json?api-key=yourkey



```python
import pandas as pd
import requests
import os
from dotenv import load_dotenv

# Load api key from .env and then insert it into the NYT API endpoint url
load_dotenv(override=True) 
api_key = os.getenv("API_KEY")
nyt_endpoint = f"https://api.nytimes.com/svc/books/v3/lists/current/hardcover-fiction.json?api-key={api_key}"


try:
    # fetch json data from the NYT API
    response = requests.get(nyt_endpoint, timeout=20)
    response.raise_for_status()
    # convert the returned json data into a dataframe and then print
    best_sellers_df = pd.DataFrame(response.json())
    print(best_sellers_df.head())
except requests.exceptions.HTTPError as err:
    print(f"HTTP error occurred: {err}")
except Exception as err:
    print(f"An error occurred: {err}")
```
